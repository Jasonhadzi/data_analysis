```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as  plt
import seaborn as sns
```

```python
all_data = pd.read_csv('all_data.csv')

print(all_data.head())
print(all_data.dtypes)
```

```python
#We change the name of the Life expectancy at birth (years) column
all_data = all_data.rename({"Life expectancy at birth (years)":"LEABY"}, axis = "columns")
print(all_data.head())
```

```python
print(all_data.describe())
```

 ### Find the average LEABY and GDP by country.

```python
all_data['Country']= all_data['Country'].replace({"United States of America":"USA"})

all_data_means = all_data.drop("Year", axis = 1).groupby("Country").mean().reset_index()
print(all_data_means)
```

```python
palette =  sns.color_palette("pastel")
sns.set_palette(palette)
sns.palplot(palette)
```

```python
sns.barplot(x="LEABY", y="Country", data=all_data_means)
plt.xlabel("Life expectancy at birth (years)")
plt.show()
```

```python
sns.barplot(x="GDP", y="Country", data=all_data_means)
plt.xlabel("GDP in Trillions of U.S. Dollars")
plt.show()
```

### Has life expectancy increased over time in the six nations?

```python
sns.lineplot(data = all_data, x='Year', y ='LEABY', hue='Country')
# Put the legend out of the figure
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.show()
```

```python
# we want to remove Zimbabwe in order to get a better understanding for the rest of the countries

countries_wo_Zimb = all_data.drop(index=all_data[all_data['Country'] == 'Zimbabwe'].index)

#print(countries_wo_Zimb['Country'].unique())
#print(all_data['Country'].unique())

sns.lineplot(data = countries_wo_Zimb, x='Year', y ='LEABY', hue='Country')
# Put the legend out of the figure
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.show()
```

### Has GDP increased over time in the six nations?

```python
sns.lineplot(data = all_data, x='Year', y ='GDP', hue='Country')
# Put the legend out of the figure
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.show()
```
